```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

bound = 100



fig = plt.figure()

ax = plt.axes(xlim=(0, bound), ylim=(-30, 5))



linebase, = ax.plot([-18]*bound,color='red',linestyle='dashed',label='Real GS')
line_virtual, = ax.plot([-18]*bound,color='blue',label='Virtual GS')
line, = ax.plot([-18]*bound,color='orange',label='Energy')


ax.legend()


virtual_gs = -18
current_e = 0
energies=[0]

def update(i):
    global virtual_gs
    global current_e
    global energies
    
    if current_e > -18:
        virtual_gs += -0.10
    else :
        virtual_gs += 0.10
    noise = (2*np.random.rand()-1)*((np.abs(current_e-virtual_gs)/8)**1)
    current_e += -(current_e-virtual_gs)/(18) + noise
    energies.append(current_e)
    
        

def animate(i):
    global virtual_gs
    global current_e
    global energies
    update(i)
#     print(np.linspace(0, i, i+1))
#     print(energies[0:i+1])
#     print('-------------')
    line.set_xdata(np.linspace(0, i, i+1))
    line.set_ydata(energies[0:i+1])  # update the data.
    line_virtual.set_ydata(virtual_gs)  # update the data.
    
    return line, line_virtual, 


ani = animation.FuncAnimation(fig, animate, interval=50, blit=True, save_count=bound)

# To save the animation, use e.g.
#
ani.save("movie.gif")


plt.show()
```
